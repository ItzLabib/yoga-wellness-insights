```python
import pandas as pd
import numpy as np

# Generate Synthetic Data
np.random.seed(42)

dates = pd.date_range(start="2024-12-01", periods=30, freq='D')
session_duration = np.random.randint(20, 60, size=30)
pose_proficiency = np.random.randint(75, 95, size=30)
engagement_score = np.random.choice([0, 1], size=30, p=[0.2, 0.8])
session_type = np.random.choice(
    ['Relaxation Yoga', 'Power Yoga', 'Hatha Yoga', 'Vinyasa Yoga', 'Restorative Yoga'], 
    size=30
)
calories_burned = session_duration * np.random.uniform(4, 6, size=30).astype(int)

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Session_Duration": session_duration,
    "Pose_Proficiency": pose_proficiency,
    "Engagement_Score": engagement_score,
    "Session_Type": session_type,
    "Calories_Burned": calories_burned
})

# Save to CSV
df.to_csv("yoga_wellness_data.csv", index=False)
print("Synthetic dataset generated and saved!")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("yoga_wellness_data.csv")

# Preview the data
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Describe the data
print(df.describe())
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Weekly summary
weekly_summary = df.resample('W', on='Date').mean()

# Display trends
print("Weekly Summary:")
print(weekly_summary)
```

```python
import matplotlib.pyplot as plt

# Plot trends
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Session_Duration'], label='Session Duration')
plt.plot(df['Date'], df['Pose_Proficiency'], label='Pose Proficiency')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Session Trends Over Time')
plt.legend()
plt.show()
```

```python
# Generate insights
avg_duration = df["Session_Duration"].mean()
avg_proficiency = df["Pose_Proficiency"].mean()

print(f"Average Session Duration: {avg_duration:.2f} mins")
print(f"Average Pose Proficiency: {avg_proficiency:.2f}%")

if avg_proficiency < 85:
    print("Focus on improving balance and alignment in poses.")
else:
    print("Great job! Keep maintaining your proficiency.")
```
